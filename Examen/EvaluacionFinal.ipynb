```python
def espita(d):
    x = []
    q,r,t,k,n,l = 1,0,1,1,3,3
    while len(x) < d:
        if 4*q+r-t < n*t:
            x.append(n)
            q,r,t,k,n,l = (
                10*q,10*(r-n*t),t,k,
                (10*(3*q+r))//t-10*n,l)
        else:
            q,r,t,k,n,l = (
                q*k,(2*q+r)*l,t*l,k+1,
                (q*(7*k+2)+r*l)//(t*l),l+2)
    return x
```

## <font color = 'orange'> Problema 5 </font>
**Similar la Barnsley fern realizada en clase**
https://en.wikipedia.org/wiki/Barnsley_fern,
elegir por lo menos otras dos opciones de https://es.wikipedia.org/wiki/Fractal  Programar y mostrar con una elección de color adecuada. 

```python
# Respuesta: 
from numba import autojit
@autojit

def fractal (re, im, max_iter):
    c = complex(re, im)
    z = 0.0j
    
    for i in range(max_iter):
        z = z*z+c
        if(z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iter

columns = 2000
rows = 2000

result = np.zeros([rows, columns])
for row_index, re in enumerate(np.linspace(-2, 1, num=rows)):
    for column_index, im in enumerate(np.linspace(-1,1, num=columns)):
        result[row_index, column_index] = fractal(re, im, 100)

plt.figure(dpi = 100)
plt.imshow(result.T, cmap = "inferno", interpolation="bilinear", extent = [-2,1,-1,1])
plt.xlabel("re")
plt.ylabel("im")
plt.show()
```

```python
from math import sin, cos, pi
import matplotlib.pyplot as plt
import matplotlib
 
def kochSnowFlake(l, n):
    x_vertex1 = 0
    y_vertex1 = 0
     
    x_vertex2 = l * cos ( 2 * pi/3)
    y_vertex2 = l * sin(2 * pi/3)
     
    x_vertex3 = l * cos (pi/3)
    y_vertex3 = l * sin(pi/3)
     
    kochCurve(x_vertex1, y_vertex1, x_vertex2, y_vertex2, n)
    kochCurve(x_vertex2, y_vertex2, x_vertex3, y_vertex3, n)
    kochCurve(x_vertex3, y_vertex3, x_vertex1, y_vertex1, n)
     
    return
         
def kochCurve(xi, yi, xf, yf, n):
    if n==0:
       plt.plot([xi,xf],[yi,yf])
             
    elif n>0:
        x1 = xi + (xf - xi) / 3.0
        y1 = yi + (yf - yi) / 3.0
 
        x3 = xf - (xf - xi) / 3.0
        y3 = yf - (yf - yi) / 3.0
         
        x2 = (x1 + x3) * cos (pi/3) - ( y3 - y1) * sin (pi/3)
        y2 = (y1 + y3) * cos (pi/3) + ( x3 - x1) * sin (pi/3) 
 
        kochCurve(xi,yi,x1,y1,n-1)
        kochCurve(x1,y1,x2,y2,n-1)
        kochCurve(x2,y2,x3,y3,n-1)
        kochCurve(x3,y3,xf,yf,n-1)
         
    return
   
plt.figure("Koch snowflack")
kochSnowFlake(200,4) # range 1 -6
plt.show()
```
